# ecdfplot

```python
sns.ecdfplot(data=df, x="10percentage",hue = "Gender", stat = "count")
```

# Violinplot

```python
sns.violinplot(data=df, x="Gender", y="10percentage")
```

# Boxenplot

```python
sns.boxenplot(data=df, x="Gender", y="10percentage")
```

# Pointplot

```python
sns.pointplot(data=df, x="Gender", y="10percentage")
```

# Heatmap

```python
sf = df.loc[:,["Salary"]]
sf
```

```python
sns.heatmap(sf, vmin=300000, vmax=2000000, cmap="vlag")
```

# lmpplot

```python
sns.lmplot(data=df, x="collegeGPA", y="10percentage", hue ="Gender", col = "Gender")
```

# FaceGrid

```python
g = sns.FacetGrid(df, col = "Gender", hue = "Gender")
g.map_dataframe(sns.scatterplot, x="10percentage", y="collegeGPA")
g.set_axis_labels("10pecentage", "collegeGPA")
```

# Pairplot

```python
sf = df.loc[:,["10percentage","collegeGPA", "12percentage","Gender"]]
```

```python
g = sns.pairplot(sf , hue = "Gender", markers = ["s","d"], diag_kind="kde")
g.map_lower(sns.kdeplot, levels=4, color=".2")
```

# Jointplot

```python
sns.jointplot(data=sf, x="10percentage", y="collegeGPA", hue = "Gender")
```

# Clustermap

```python
sf = df.loc[:,["10percentage","collegeGPA", "12percentage", "Gender"]]
```

```python
Gender = sf.pop("Gender")
lut = dict(zip(Gender.unique(), "rbg"))
row_colors = Gender.map(lut)
g = sns.clustermap(sf, row_colors=row_colors)
```
